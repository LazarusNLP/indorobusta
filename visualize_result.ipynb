# Plot Accuracy vs Perturbation By Language

6 Bahasa = 6 Figures
Tiap figures punya 4 linechart untuk tiap model

### Sentiment - English

```python
# plt.style.use('ggplot')

def visualize_1(lang, task, df=df_comb):   
    lang_map = {
        "en": "English",
        'fr': "French", 
        'id': "Indonesian", 
        'it': "Italian", 
        'jw': "Javanese", 
        'ms': "Malay", 
        'su': "Sundanese"
    }

    color_map = {
        'indobert': "#f5c63c", 
        'indobertlarge': "#f47f68", 
        'xlmrlarge': "#b422c7",
        'mbert': "#1c348c", 
        'xlmr': "#2a9fe8"
    }

    task_map = {
        "emotion": "Emotion Classification",
        "sentiment": "Sentiment Classification"
    }


    df_lang = df[(df["codemix_lang"] == lang) & (df["downstream_task"] == task)]

    percent_pert = [0, 0.2, 0.4, 0.6, 0.8]
    list_model = df_lang.model_target.unique()
    
    figure(num=None, figsize=(8, 6), dpi=80)


    for model in list_model:
        acc_arr = df_lang[(df_lang["model_target"] == model)].after_attack_acc.values
        before_attack_acc_arr = df_lang[(df_lang["model_target"] == model)].before_attack_acc.values[0]
        acc_arr = np.insert(acc_arr, 0, before_attack_acc_arr, axis=0)
        
        for i, v in enumerate(acc_arr):
            plt.annotate(str(format(float(v),".2f")), xy=(percent_pert[i],v), textcoords='offset points', xytext=(5, 5))
        
        if "large" in model:
            plt.plot(percent_pert, acc_arr, '-o', label='{}'.format(model), linestyle='--', color=color_map[model])
        else:
            plt.plot(percent_pert, acc_arr, '-o', label='{}'.format(model), color=color_map[model])
        # plt.plot(percent_pert, acc_arr, '-o', label='{}'.format(model), color=color_map[model])
    
    plt.legend(loc="lower left")
    plt.xlabel('Perturbation Ratio')
    plt.ylabel('Accuracy Score')
    title = 'Indonesian-{} {}'.format(lang_map[lang], task_map[task])
    plt.title(title)
    
    # plt.savefig('plot.png')
    
    plt.savefig('viz/rq1/'+str(title)+'.png')
    plt.show()

# lang_list = ['en', 'fr', 'id', 'it', 'jw', 'ms', 'su']
# for i in lang_list:
task = "emotion"
visualize_1("en", task)
visualize_1("fr", task)
visualize_1("id", task)
visualize_1("it", task)
visualize_1("jw", task)
visualize_1("ms", task)
visualize_1("su", task)

# anotasi bisa diganti warnanya sesuai warna linechart
# Kalau judul kepanjangan bisa disingkat pake lang_code

# seed24032022
# indobert-emotion-codemixing-en-adv-0.6-valid
# mbert-emotion-sr-adv-0.6-valid
# xlmr-emotion-sr-adv-0.6-valid
# indobert-emotion-codemixing-jw-adv-0.4-valid
# mbert-emotion-codemixing-jw-adv-0.4-valid
# indobert-emotion-codemixing-jw-adv-0.6-valid
# mbert-emotion-codemixing-jw-adv-0.6-valid
# indobert-emotion-codemixing-jw-adv-0.8-valid
# mbert-emotion-codemixing-jw-adv-0.8-valid
# mbert-emotion-codemixing-ms-adv-0.4-valid
# mbert-emotion-codemixing-su-adv-0.4-valid

# seed42
# xlmr-sentiment-codemixing-en-adv-0.6-valid
# xlmr-sentiment-codemixing-jw-adv-0.6-valid
```

```python
# plt.style.use('ggplot')

def visualize_2(lang, task, df=df_comb):   
    lang_map = {
        "en": "English",
        'fr': "French", 
        'id': "Indonesian", 
        'it': "Italian", 
        'jw': "Javanese", 
        'ms': "Malay", 
        'su': "Sundanese"
    }

    color_map = {
        'indobert': "#f5c63c", 
        'indobertlarge': "#f47f68", 
        "xlmrlarge": "#b422c7",
        'mbert': "#1c348c", 
        'xlmr': "#2a9fe8"
    }

    task_map = {
        "emotion": "Emotion Classification",
        "sentiment": "Sentiment Classification"
    }


    df_lang = df[(df["codemix_lang"] == lang) & (df["downstream_task"] == task)]

    percent_pert = [0, 0.2, 0.4, 0.6, 0.8]
    list_model = df_lang.model_target.unique()
    
    figure(num=None, figsize=(8, 6), dpi=80)


    for model in list_model:
        acc_arr = df_lang[(df_lang["model_target"] == model)].after_attack_acc.values
        before_attack_acc_arr = df_lang[(df_lang["model_target"] == model)].before_attack_acc.values[0]
        acc_arr = np.insert(acc_arr, 0, before_attack_acc_arr, axis=0)
        semantic_sim_arr = df_lang[(df_lang["model_target"] == model)].avg_semantic_sim.values
        semantic_sim_arr = np.insert(semantic_sim_arr, 0, 1, axis=0)

        for i, v in enumerate(acc_arr):
            plt.annotate(str(format(float(v),".2f")), xy=(semantic_sim_arr[i],v), textcoords='offset points', xytext=(5, 5))
        
        
        if "large" in model:
            plt.plot(semantic_sim_arr, acc_arr, '-o', label='{}'.format(model), linestyle='--', color=color_map[model])
        else:
            plt.plot(semantic_sim_arr, acc_arr, '-o', label='{}'.format(model), color=color_map[model])
        # plt.plot(semantic_sim_arr, acc_arr, '-o', label='{}'.format(model), color=color_map[model])
    
    plt.legend(loc="lower left")
    plt.xlabel('Semantic Similarity')
    plt.ylabel('Accuracy Score')
    title = 'Indonesian-{} {}'.format(lang_map[lang], task_map[task])
    plt.title(title)
    
    # plt.savefig('plot.png')
    
    plt.savefig('viz/rq2/'+str(title)+'.png')
    plt.show()

# lang_list = ['en', 'fr', 'id', 'it', 'jw', 'ms', 'su']
# for i in lang_list:
task = "emotion"
visualize_2("en", task)
visualize_2("fr", task)
visualize_2("id", task)
visualize_2("it", task)
visualize_2("jw", task)
visualize_2("ms", task)
visualize_2("su", task)
```
